# TxMineFlow - exploration
Queries the DuckDB warehouse built by the pipeline. This notebook is the 'notebook layer' (BGCFlow pattern); `txmineflow report` renders it to HTML.

```
# parameters (papermill-injectable)
db_path = 'results/warehouse/txmineflow.duckdb'
padj_cut = 0.05
```

```
import duckdb, pandas as pd
con = duckdb.connect(db_path, read_only=True)
de = con.execute('SELECT * FROM differential_expression').df()
print(de.shape); de.head()
```

## Volcano plot

```
import numpy as np, matplotlib.pyplot as plt
de['sig'] = de['padj'] < padj_cut
plt.figure(figsize=(6,5))
plt.scatter(de['log2FoldChange'], -np.log10(de['pvalue']),
            c=de['sig'].map({True:'crimson', False:'lightgray'}), s=8)
plt.xlabel('log2 fold change'); plt.ylabel('-log10 p')
plt.title('Differential expression'); plt.tight_layout()
```
